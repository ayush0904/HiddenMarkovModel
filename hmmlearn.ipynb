```python
import sys
import math
```

### Emission and Transition  Probability

```python
nameOfFile = 'hmm-training-data/it_isdt_train_tagged.txt'
with open(nameOfFile) as fp:
    train_data = fp.readlines()
```

```python
emission_probability_dict = {}
transition_probability_dict = {}
```

```python
for data in train_data:
    tokens = data.split()
    start_token = "<start>" 
    for element in tokens:
        #the separator is always the last slash in the word/tag sequence
        word, tag = element.rsplit("/", 1)
        
        #This has to be updated
        if start_token not in transition_probability_dict:
            transition_probability_dict[start_token] = {}
        if tag not in transition_probability_dict[start_token]:
            transition_probability_dict[start_token][tag] = 1
        else:
            transition_probability_dict[start_token][tag] = transition_probability_dict[start_token][tag] + 1
        
        #This is Correct
        if tag not in emission_probability_dict:
            emission_probability_dict[tag] = {}
        if word not in emission_probability_dict[tag]:
            emission_probability_dict[tag][word] = 1 
        else:
             emission_probability_dict[tag][word] =  emission_probability_dict[tag][word] + 1
                
        start_token = tag
        
    #Check for this Portion
    if start_token not in transition_probability_dict:
        transition_probability_dict[start_token] = {}
    if "<end>" not in transition_probability_dict[start_token]:
        transition_probability_dict[start_token]["<end>"] = 1
    else:
         transition_probability_dict[start_token]["<end>"] = transition_probability_dict[start_token]["<end>"] + 1
```

```python
list_of_tags = list(transition_probability_dict.keys()) + ["<end>"]
```

### Smoothing for transaction probability

```python
for tag in transition_probability_dict: 
    total_events = 0
    for n_tag in list_of_tags:
        if n_tag not in transition_probability_dict[tag]:
            transition_probability_dict[tag][n_tag] = 1
        else:
            transition_probability_dict[tag][n_tag] += 1
        total_events += transition_probability_dict[tag][n_tag]

    for n_tag in transition_probability_dict[tag]:
        transition_probability_dict[tag][n_tag] = math.log(transition_probability_dict[tag][n_tag]/total_events)
```

### Emission Probobility

```python
for tag in emission_probability_dict:
    total_events = 0
    for word in emission_probability_dict[tag]:
        total_events += emission_probability_dict[tag][word]
    for word in emission_probability_dict[tag]:
        emission_probability_dict[tag][word] = math.log(emission_probability_dict[tag][word]/total_events)
```

```python
# Collect as string
# Change this Format
out_str = ""

for tag in transition_probability_dict:
	out_str += "{} {}".format(tag, "T")
	for n_tag in transition_probability_dict[tag]:
		out_str += " {}/{} ".format(n_tag,transition_probability_dict[tag][n_tag])
	out_str += "\n"

for tag in emission_probability_dict:
	out_str += "{} {}".format(tag, "E")
	for word in emission_probability_dict[tag]:
		out_str += " {}/{}".format(word,emission_probability_dict[tag][word])
	out_str += "\n"

out_str = out_str[:-1]
```

```python
with open("./hmmmodelfirst.txt", "w") as fp:
	fp.write(out_str)
```
